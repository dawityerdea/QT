```python
def solve_sc_m(initial_guess):
  J0 = 10
  kT = 0.6 * J0
  m = initial_guess
  y = np.tanh(J0*m/kT )
  return(y)

initial_guess = 0.9
y = solve_sc_m(initial_guess)
approx = abs(y - initial_guess)
counter = 1
while approx > 0.0001 and counter < 1e8:
  approx = abs(y - solve_sc_m(y))
  y = solve_sc_m(y)
  counter += 1
print('solution for an intial guess of m = 0.2 is m =', y)

initial_guess = -0.00001
y = solve_sc_m(initial_guess)
approx = abs(y - initial_guess)
counter = 1
while approx > 0.0001 and counter < 1e8:
  approx = abs(y - solve_sc_m(y))
  y = solve_sc_m(y)
  counter += 1
print('solution for an intial guess of m = -0.00001 is m =', y)

initial_guess = -0.7
y = solve_sc_m(initial_guess)
approx = abs(y - initial_guess)
counter = 1
while approx > 0.0001 and counter < 1e8:
  approx = abs(y - solve_sc_m(y))
  y = solve_sc_m(y)
  counter += 1
print('solution for an intial guess of m = 0.7 is m =', y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants in SI units
hbar = 1.0545718e-34  # Reduced Planck constant (J·s)
k_B = 1.380649e-23    # Boltzmann constant (J/K)
omega = 1e13          # Angular frequency (rad/s)
N = 6e23              # Number of particles (e.g., for 1 mole of atoms)

# Define beta and the specific heat function
beta = lambda T: 1 / (k_B * T)  # Beta as a function of T

def C_v(T):
    b = beta(T)
    x = b * hbar * omega / 2
    return 3 * N * k_B * (x**2) / (np.sinh(x)**2)

# Temperature range
T = np.linspace(1, 400, 1000)  # Temperatures in Kelvin

# Compute specific heat
C_v_values = C_v(T)

# Gas constant R
R = N * k_B  # J/(mol*K)


# Plot
plt.figure(figsize=(8, 6))
plt.plot(T, C_v_values, label=r"$C_V(T)$", color='blue')
plt.xlabel(r"Temperature $T$ (K)", fontsize=12)
plt.ylabel(r"Specific Heat $C_V(T)$ (J/K)", fontsize=12)
plt.grid(True)
plt.axhline(y=3*R, color='k', linestyle='-', label=f"$R = {R:.2e} \, \\mathrm{{J/(mol \cdot K)}}$")
plt.show()
```
